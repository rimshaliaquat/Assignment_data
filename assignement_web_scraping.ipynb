```python
import pandas as pd
from bs4 import BeautifulSoup 
import requests
```

```python
page=requests.get("https://en.wikipedia.org/wiki/Web_scraping")
page.status_code
```

```python
soup=BeautifulSoup(page.content, "html.parser")
# if u want to remove the space between text and put _ i.e., afshan liaquat.....afshan_liaquat
#then write   data['Names']=data['Names'].str.replace(" ","_")
```

# For h1:

```python
title= soup.find("h1").text
print(title)
```

# For Div:

```python
div= soup.find("script").text
print(div)
```

# For Footers:

```python
footer= soup.find("footer").text
print(footer)
```

# For a:

```python
a= soup.find("a").text
print(a)
//nothing output because no text outside a
```

# For Body:

```python
body= soup.find("body").text
print(body)
```

# For html tag:


```python
html= soup.find("html").text
print(html)
```
